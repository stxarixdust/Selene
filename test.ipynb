# Luna's Additive Synth

A very basic oscillator model for an additive synth.

## Function definition

This is the definition of the oscillator's quite simple funtion. In essence, the
function returns the sum of all overtones of the base frequency $f$.

$$
y(x)_{m,o,f} = \sum_{n=1}^{m}\frac{1}{n^{o}}\cdot\sin\left(x\cdot n\cdot\pi\cdot f\right)
$$

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def y(m, o, f):
  gr = 200
  x = np.linspace(-1,1,gr)
  y = np.zeros(gr)
  for n in range(1, m):
    i = 0
    while i < gr:
      y[i] += (1 / n**o) * math.sin(x[i] * n * math.pi * f)
      i += 1

  # setting the axes at the centre
  fig = plt.figure()
  ax = fig.add_subplot(1, 1, 1)
  ax.spines['left'].set_position('center')
  ax.spines['bottom'].set_position('zero')
  ax.spines['right'].set_color('none')
  ax.spines['top'].set_color('none')
  ax.xaxis.set_ticks_position('bottom')
  ax.yaxis.set_ticks_position('left')

  # plot the function
  plt.plot(x,y, 'r')

  # show the plot
  plt.show()
```

Here is a basic example of usage of this function:

```python
y(8,1,4)
```

## Parameters

This section goes into detail about the two parameters of the aforementioned
function and showcases how they shape the sound.

$o\geq1$ is the dampening factor. If it is set to 1, the amplitudes of the
overtones are scaled to produce a saw wave. If it is greater than 1, the
amplitudes decrease more significantly, smoothing the resulting signal and
bringing it closer to a sine. The greater it is, the softer and less harsh the
resulting signal sounds.

This is the previous example with more intense dampening. Notice how the signal
is now more sinusoidal, as emphasis is removed from higher overtones, and moved
towards the base tone.

```python
y(8,1.6,4)
```

$m > 0\ \epsilon\ \Z$ is the harmonic count. It dictates how many overtones
the signal has. Especially on low dampening, the difference between values
for this parameter becomes clear.

This is the first example with more overtones.

```python
y(32,1,4)
```
